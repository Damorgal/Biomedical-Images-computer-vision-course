El grafo de conectividad es una matriz traingular superior donde cada entrada $i,j$ representa el peso de un arista bidireccional entre los nodos (que representan partes del cerebro) $i$ y $j$.

```python
# Convertimos a arreglo numpy para trabajar mejor con el
data_connections = np.array(data_connections)
data_distance = np.array(data_distance)
```

Visualizamos:

```python
# Recordamos que es una matriz triangular superior
data = data_connections.copy()
diag = np.diag(np.diag(data))
data_sim_con = data.copy()
data_sim_con += data.T
data_sim_con -= diag
data = data_distance.copy()
diag = np.diag(np.diag(data))
data_sim_dis = data.copy()
data_sim_dis += data.T
data_sim_dis -= diag
```

```python
plt.figure(figsize=(20, 10))
plt.subplot(121)
plt.title('Conexiones (Cantidad) de Streamlines entre Nodos')
plt.imshow(data_sim_con, 'plasma')
plt.axis('off')

plt.subplot(122)
plt.title('Distancia (Longitud Media) de los Streamlines entre Nodos')
plt.imshow(data_sim_dis, 'plasma')
plt.axis('off')
plt.show()
```

### Métrica básica: Vertex Degree.

En esta métrica de las más básicas, se cuenta el número de aristas que salen de cada nodo de conectividad. Si no hay una conexión entre nodos, se sustituye por peso 0. Se selecciona esta métrica debido a que a pesar de ser una de las básicas, es muy usada para algunos teoremas de partición de grafos.

```python
def vertexDegree(G):
    degree = np.zeros(G.shape[0], dtype=int)
    for i in range(G.shape[0]):
        count = 0
        for j in range(i, G.shape[1]):
            if G[i, j]:
                degree[i] += 1
                if i != j:
                    degree[j] += 1
    return degree
```

Para este caso, no importa que datos usemos, si los de streamlines o los de distancia media entre los nodos, puesto que si existe una arista entre dos nodos, ésta estará en los dos grafos ya sea por medio de streamlines o de distancia:

Pobamos la métrica a continuación:

```python
VD = vertexDegree(data_connections)
VD_df = pd.DataFrame(VD, columns=['Vertex Degree']).T
VD_df
```

### Métrica de distancia: Graph Distance.

Para ahora probar una métrica más compleja, probaremos con las distancias del grafo, es decir, el camino más corto entre un nodo $i$ y un nodo $j$. Para este caso, podemos correr el algoritmo para el grafo de conectividad pesado en cantidad de streamlines o en la longitud media de ellos, que para este caso, sería más interesante realizarlo con la longitud media de cada streamlines pues esto nos dará una idea de la longitud más corta entre una zona del cerebro y otra.

Para proceder a calcular ésta métrica, se necesita programar un algoritmo para encontrar el camino más corto entre dos nodos. Aquí se selecciona el método de Dijkstra para proceder a ello.

```python
# Instalación:
#!pip install dijkstra
```

Creamos el grafo de conexión para la clase Dijkstra que usaremos:

```python
def createGraph(G):
    graph = Graph()
    for i in range(G.shape[0]):
        count = 0
        for j in range(i, G.shape[1]):
            if G[i, j]:
                graph.add_edge(i, j, G[i, j])
                # Porque es bidireccional
                if i != j:
                    graph.add_edge(j, i, G[i, j])
    return graph
```

```python
graph = createGraph(data_distance)
```

Ahora para cada nodo, calculamos el camino más corto y creamos la matriz del grafo de distancia:

```python
def getGraphDistance(data, graph):
    graph_dist = []
    for i in range(data.shape[0]):
        # Calculamos el camino más corto desde el nodo i
        dijkstra = DijkstraSPF(graph, i)
        distances = [] 
        for j in range(data.shape[1]):
            distances.append(dijkstra.get_distance(j))
        graph_dist.append(distances)
    return np.array(graph_dist)
```

```python
GD = getGraphDistance(data_distance, graph)
GD_pd = pd.DataFrame(GD)
```

Visualizamos los datos. Notar que es una matriz simétrica pues el camino más corto del nodo $i$ al nodo $j$ es el mismo debido a que es un grafo bidireccional.

```python
GD_pd.head()
```

```python
plt.figure(figsize=(20, 10))
plt.title('Grafo de Distancia (Camino más Corto entre Nodos)')
plt.imshow(GD, 'plasma')
plt.axis('off')
plt.show()
```
